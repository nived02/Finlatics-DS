# **What** is the distribution of age among the clients?

```python
import pandas as pd
df=pd.read_csv(r"/content/banking_data.csv")
```

```python
print(df.head())
```

```python
print(df['age'].count)
```

```python
print(df['age'].max())
print(df['age'].min())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=20, edgecolor='black')
plt.title('Age Distribution of Clients')
plt.xlabel('Age')
plt.ylabel('Number of Clients')
plt.grid(True)
plt.show()
```

# **What is the distribution of the number of days passed since the client was last contacted from a previous campaign?**

```python
print(df['pdays'].value_counts().sort_index())
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of the 'pdays' column
plt.figure(figsize=(10, 6))
plt.hist(df['pdays'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Number of Days Since Last Contact from Previous Campaign')
plt.xlabel('Number of Days')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```
